```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()
```

```python
# Let's check the description of the dataset 

print(boston.DESCR) 
```

```python
print(boston.feature_names)
```

```python
print(boston.target )
```

```python
df = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
df.head()
```

```python
df['price'] = boston.target
```

```python
df.info( )
```

```python
##  summary
df.describe()
```

```python
# Check Missing
df.isnull().sum()
```

# Exploratory Data Analysis

```python
# Correlation 

df.corr()
```

```python
import seaborn as sns
sns.pairplot(df) 
```

```python
sns.regplot(x = 'RM', y = 'price' ,data = df )
```

```python
sns.regplot(x='CRIM', y = 'price', data = df)
```

```python
sns.regplot(x='PTRATIO', y = 'price', data = df)
```

```python
## Independent and Dependent Features
X = df.drop('price', axis = 1)
y = df['price']
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
## Standardizing the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Model Training

```python
from sklearn.linear_model import LinearRegression 
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
# print the coefficient and intercepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
```

```python
plt.scatter(y_test, reg_pred)
```

```python
# residuals

residuals = y_test - reg_pred
```

```python
# plotting the residuals
# Showing normal distribution indicates good performance
sns.displot(residuals, kind = 'kde')
```

```python
## Scatter plot with respect to predictions and residuals
# Showing uniform distribution indicates good performance
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))

print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# R Square and Adjusted R square
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
# Adjusted R square = 1 - [(1-R2) * (n-1)/(n-k-1)]
1 - (1-score) * (len(y_test)-1)/(len(y_test) - X_test.shape[1] -1) 
```
